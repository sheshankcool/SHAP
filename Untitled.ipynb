```python
import shap
import sklearn
import xgboost
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
shap.initjs()
```

```python
X,y = shap.datasets.adult()
```

```python
X.shape
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=33,stratify=y,test_size=.14)
```

```python
print(X_train.shape)
X.head()
```

```python

model = xgboost.XGBClassifier()
model.fit(X_train,y_train)
```

```python
explainer=shap.TreeExplainer(model,X_train,model_output="log_loss")
```

```python
shap_values=explainer.shap_values(X_train,y_train)
```

```python
explainer.expected_value(0)
```

```python
shap.monitoring_plot(5, shap_values, X_train)
```
